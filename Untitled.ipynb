# **Implementation of a PID Controller**

This notebook explores the theory and implementation of a **PID (Proportional-Integral-Derivative)** controller. PID control is the most common control algorithm used in industry (robotics, thermostats, cruise control) because of its simplicity and effectiveness.

---

## **1. The PID Equation**
The controller calculates an "error" value $e(t)$ as the difference between a measured process variable and a desired setpoint. The controller attempts to minimize the error by adjusting a control input $u(t)$.

The standard form of the controller is:

$$u(t) = K_p e(t) + K_i \int_{0}^{t} e(\tau) d\tau + K_d \frac{de(t)}{dt}$$

### **The Three Components:**
* **Proportional ($K_p$):** Produces an output value that is proportional to the current error. If the error is large, the control output is large.
* **Integral ($K_i$):** Accounts for past values of the error. If the error has been persisting for a long time, the integral term increases to eliminate the residual steady-state error.
* **Derivative ($K_d$):** Predicts future error based on its current rate of change. It acts as a "brake," slowing down the correction as the setpoint is approached to prevent overshoot.

---

## **2. Discrete Implementation**
Since computers work in discrete time steps ($\Delta t$), we approximate the integral and derivative:

1.  **Integral:** $\text{integral} = \text{integral} + (e(t) \cdot \Delta t)$
2.  **Derivative:** $\text{derivative} = \frac{e(t) - e(t-1)}{\Delta t}$

## **3. The Python Class**
*Run the code cell below to initialize our PID controller.*

```python
import numpy as np
import matplotlib.pyplot as plt

class PID:
    def __init__(self, Kp, Ki, Kd, setpoint=0):
        self.Kp = Kp
        self.Ki = Ki
        self.Kd = Kd
        self.setpoint = setpoint
        
        self.prev_error = 0
        self.integral = 0

    def update(self, measurement, dt):
        error = self.setpoint - measurement
        
        # Proportional term
        P = self.Kp * error
        
        # Integral term
        self.integral += error * dt
        I = self.Ki * self.integral
        
        # Derivative term
        derivative = (error - self.prev_error) / dt
        D = self.Kd * derivative
        
        self.prev_error = error
        return P + I + D
```
