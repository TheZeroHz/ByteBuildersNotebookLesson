
# 📡 Lesson: WiFi Scanning on ESP32-S3 (MicroPython)

## 🧠 Overview
This lesson teaches how to **scan nearby WiFi networks** using an ESP32-S3 running MicroPython. Scanning helps you:

- Detect available WiFi networks
- Check signal strength (RSSI)
- Identify security type
- Obtain network channels and BSSID (router MAC address)

The ESP32 must be in **Station Mode (STA)** to scan.

---

## 🔧 Required Modules

```python
import network
import time
````

* **`network`**: Provides WiFi control (Station/Access Point, connect, scan, etc.)
* **`time`**: Used for delays or time-based operations

---

## ⚙️ Step 1: Initialize WiFi Interface

```python
wlan = network.WLAN(network.STA_IF)
wlan.active(True)
time.sleep(1)  # optional, ensures WiFi is ready
```

* `network.WLAN(network.STA_IF)` → Creates **station interface** (for connecting/scanning networks)
* `wlan.active(True)` → Powers on the WiFi module

---

## 🔍 Step 2: Scan for Networks

```python
networks = wlan.scan()
```

* Returns a **list of tuples**, each tuple representing a network:

```text
(ssid, bssid, channel, rssi, authmode, hidden)
```

| Index | Description                |
| ----- | -------------------------- |
| 0     | SSID (WiFi name)           |
| 1     | BSSID (MAC address)        |
| 2     | Channel (1–13)             |
| 3     | RSSI (signal strength dBm) |
| 4     | Authmode (security type)   |
| 5     | Hidden (True/False)        |

---

## ⚠️ Step 3: Handle No Networks

```python
if not networks:
    print("No networks found.")
    return
```

* Checks if any networks are detected
* If empty, prints a message and exits function

---

## 🔁 Step 4: Iterate Over Networks

```python
for i, net in enumerate(networks):
```

* Loops through all detected networks
* `i` → index number for display
* `net` → the tuple of network info

---

## 🏷️ Step 5: Extract Network Information

```python
ssid = net[0].decode('utf-8')
bssid = ':'.join('{:02x}'.format(b) for b in net[1])
channel = net[2]
rssi = net[3]
authmode = net[4]
hidden = net[5]
```

* `ssid` → Convert bytes to string
* `bssid` → MAC address formatted as `xx:xx:xx:xx:xx:xx`
* `channel` → WiFi channel
* `rssi` → Signal strength in dBm
* `authmode` → Numeric code for security type
* `hidden` → Whether SSID is hidden

---

## 🔐 Step 6: Convert Auth Mode

```python
def get_auth_mode(mode):
    modes = {
        0: "Open",
        1: "WEP",
        2: "WPA-PSK",
        3: "WPA2-PSK",
        4: "WPA/WPA2-PSK"
    }
    return modes.get(mode, "Unknown")
```

* Maps numeric codes to human-readable security names
* Example:

  * `0` → Open (no password)
  * `3` → WPA2-PSK (common home WiFi)

---

## 🖨️ Step 7: Print Network Info

```python
print(f"{i+1}. SSID     : {ssid}")
print(f"   BSSID    : {bssid}")
print(f"   Channel  : {channel}")

`````python
import network
import time

def scan_wifi():
    wlan = network.WLAN(network.STA_IF)
    wlan.active(True)

    print("Scanning for WiFi networks...\n")
    
    networks = wlan.scan()  # returns list of tuples

    if not networks:
        print("No networks found.")
        return

    for i, net in enumerate(networks):
        ssid = net[0].decode('utf-8')
        bssid = ':'.join('{:02x}'.format(b) for b in net[1])
        channel = net[2]
        rssi = net[3]
        authmode = net[4]
        hidden = net[5]

        print(f"{i+1}. SSID     : {ssid}")
        print(f"   BSSID    : {bssid}")
        print(f"   Channel  : {channel}")
        print(f"   RSSI     : {rssi} dBm")
        print(f"   Security : {get_auth_mode(authmode)}")
        print(f"   Hidden   : {hidden}")
        print("-" * 40)

def get_auth_mode(mode):
    modes = {
        0: "Open",
        1: "WEP",
        2: "WPA-PSK",
        3: "WPA2-PSK",
        4: "WPA/WPA2-PSK"
    }
    return modes.get(mode, "Unknown")

# Run scan
scan_wifi()
`````

# 📡 Lesson: WiFi Connection on ESP32-S3 (MicroPython)

## 🧠 Overview
This lesson shows how to **connect your ESP32-S3 to a WiFi network** using MicroPython. After connecting, you can:

- Use internet features (HTTP requests, MQTT, APIs)
- Get your ESP32’s IP, subnet, gateway, and DNS
- Handle connection timeouts

---

## 🔧 Required Modules

```python
import network
import time

`````python
# New code cell
def connect_wifi(ssid, password, timeout=15):
    wlan = network.WLAN(network.STA_IF)   # create station interface
    wlan.active(True)                      # enable WiFi
    time.sleep(1)                          # wait a moment for hardware
    
    # Check if already connected
    if wlan.isconnected():
        print("Already connected!")
        print("Network config:", wlan.ifconfig())
        return wlan
    
    print(f"Connecting to {ssid}...", end="")
    wlan.connect(ssid, password)           # attempt connection
    
    # Timeout handling using ticks_ms
    start = time.ticks_ms()
    while not wlan.isconnected():
        if time.ticks_diff(time.ticks_ms(), start) > timeout * 1000:
            wlan.active(False)
            raise OSError("WiFi connection timed out")
        print(".", end="")
        time.sleep(0.5)
    
    # Connected successfully
    print(" connected!")
    ip, subnet, gateway, dns = wlan.ifconfig()
    print(f"  IP      : {ip}")
    print(f"  Subnet  : {subnet}")
    print(f"  Gateway : {gateway}")
    print(f"  DNS     : {dns}")
    
    return wlan
`````

`````python
# New code cell
SSID = 'Rakib'          # replace with your WiFi name
PASSWORD = 'rakib@2025' # replace with your WiFi password
`````

`````python
# New code cell
wlan = connect_wifi(SSID, PASSWORD)
`````
